```python
import math
import matplotlib.pyplot as plt
import numpy as np
def biseksi(fx):
    a = float(input("Masukan a: "))
    b = float(input("Masukan b: "))
    c = a 
    x = np.arange(3,6,0.01)
    y = np.sin(10*x) + np.cos(3*x)  
    plt.plot(x,y,'-')
    plt.grid(True)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show() 
    print ("a\tb\tfx(a)\tfx(b)\tc\tfx(c)")
    if(fx(a)*fx(b) < 0):
        while (abs(fx(a)*fx(b)) > 0.01):
            c = (a+b)/2
            print (f'{a:.3f}\t{b:.3f}\t{fx(a):.3f}\t{fx(b):.3f}\t{c:.3f}\t{fx(c):.3f}')
            e = abs(((c-b)/c)*100/100)<0.5/100 #Menggunakan Galat relatif
            if abs(fx(c)<e):
                return c
            if(fx(a)*fx(c)>0):
                a = c
            else:
                b = c 
            # print(c)    
        print (f'Akar persamaan adalah',c) 
        print(f'{abs(e)}')
    else:
        print ('Tebakan Awal a dan b salah')
fx = lambda x: np.sin(10*x) + np.cos(3*x)  
biseksi(fx)     
```
